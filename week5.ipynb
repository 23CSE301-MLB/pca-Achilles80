```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
df = pd.read_csv("WineQT.csv")
df
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data = pd.DataFrame(scaled_data,columns=df.columns)
scaled_data
```

```python
n = 1142
sns.pairplot(scaled_data)
plt.suptitle("Pairplot of Original Features", y=1.02)
plt.show()
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(pca_data,columns=scaled_data.columns)
pca_df
```

```python
print(pca.explained_variance_ratio_)
print("\nExplained Variance (Eigen values):")
print(pca.explained_variance_)
print("\nPrincipal Axes (components/Eigen Vectors):")
print(pca.components_)
print("\nMean of each feature before transformation:")
print(pca.mean_)
```

```python
explained_var = pca.explained_variance_ratio_
plt.figure(figsize=(8,5))
plt.bar(range(1, 14), explained_var, alpha=0.7, align='center', label='Individual explained variance')
plt.step(range(1, 14), np.cumsum(explained_var), where='mid', label='Cumulative explained variance', 
color='red')
plt.xlabel('Principal Component Index')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(pca_df['fixed acidity'], pca_df['volatile acidity'], c='blue', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - 2D Projection')
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df['fixed acidity'], pca_df['volatile acidity'], pca_df['citric acid'], c='green', alpha=0.7)
# ax.set_xlabel('PC1')
# ax.set_ylabel('PC2')
# ax.set_zlabel('PC3')
ax.set_title('PCA - 3D Projection')
plt.show()
```
